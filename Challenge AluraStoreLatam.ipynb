**CÓDIGO Y RESULTADOS DE LA EVALUACIÓN DE PARÁMETROS EN LAS TIENDAS**

```python
# Importar librerías básicas
import pandas as pd

# URLs de los archivos (ya las proporcionaste)
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Lista de DataFrames para procesar todas las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# --- 1. Facturación Total ($) ---
print("1. Facturación Total por Tienda:")
facturaciones = []
for i in range(len(tiendas)):
    # Sumar todos los valores de la columna 'Precio'
    total = 0
    for precio in tiendas[i]['Precio']:
        total = total + precio
    facturaciones.append(total)
    print(f"{nombres_tiendas[i]}: ${total:,.0f}")

# --- 2. Categorías más populares ---
print("\n2. Categorías más populares por Tienda:")
for i in range(len(tiendas)):
    # Crear un diccionario para sumar precios por categoría
    categorias = {}
    for index in range(len(tiendas[i])):
        categoria = tiendas[i]['Categoría del Producto'][index]
        precio = tiendas[i]['Precio'][index]
        if categoria in categorias:
            categorias[categoria] = categorias[categoria] + precio
        else:
            categorias[categoria] = precio

    # Convertir diccionario a lista de tuplas (categoría, total)
    lista_categorias = []
    for categoria, total in categorias.items():
        lista_categorias.append((categoria, total))

    # Ordenar categorías por total (de mayor a menor) usando un bucle simple
    for j in range(len(lista_categorias)):
        for k in range(j + 1, len(lista_categorias)):
            if lista_categorias[j][1] < lista_categorias[k][1]:
                # Intercambiar posiciones
                lista_categorias[j], lista_categorias[k] = lista_categorias[k], lista_categorias[j]

    # Tomar las 3 primeras categorías
    top_categorias = lista_categorias[:3]
    print(f"{nombres_tiendas[i]}: ", end="")
    for j in range(len(top_categorias)):
        if j > 0:
            print(", ", end="")
        print(top_categorias[j][0], end="")
    print()

# --- 3. Promedio de Calificación de clientes ---
print("\n3. Promedio de Calificación por Tienda:")
calificaciones_promedio = []
for i in range(len(tiendas)):
    # Sumar todas las calificaciones
    suma = 0
    conteo = 0
    for calificacion in tiendas[i]['Calificación']:
        suma = suma + calificacion
        conteo = conteo + 1
    # Calcular promedio
    promedio = suma / conteo
    calificaciones_promedio.append(promedio)
    print(f"{nombres_tiendas[i]}: {promedio:.2f}")

# --- 4. Productos más y menos vendidos ---
print("\n4. Productos más y menos vendidos por Tienda:")
for i in range(len(tiendas)):
    # Crear un diccionario para sumar precios por producto
    productos = {}
    for index in range(len(tiendas[i])):
        producto = tiendas[i]['Producto'][index]
        precio = tiendas[i]['Precio'][index]
        if producto in productos:
            productos[producto] = productos[producto] + precio
        else:
            productos[producto] = precio

    # Convertir diccionario a lista de tuplas (producto, total)
    lista_productos = []
    for producto, total in productos.items():
        lista_productos.append((producto, total))

    # Ordenar productos por total (de mayor a menor)
    for j in range(len(lista_productos)):
        for k in range(j + 1, len(lista_productos)):
            if lista_productos[j][1] < lista_productos[k][1]:
                lista_productos[j], lista_productos[k] = lista_productos[k], lista_productos[j]

    # Productos más vendidos (top 3)
    print(f"{nombres_tiendas[i]} - Productos más vendidos:")
    for j in range(min(3, len(lista_productos))):
        print(f"  {lista_productos[j][0]}: ${lista_productos[j][1]:,.0f}")

    # Productos menos vendidos (últimos 3)
    print(f"{nombres_tiendas[i]} - Productos menos vendidos:")
    for j in range(-1, -min(4, len(lista_productos)) - 1, -1):
        print(f"  {lista_productos[j][0]}: ${lista_productos[j][1]:,.0f}")

# --- 5. Costo promedio del envío ($) ---
print("\n5. Costo promedio del envío por Tienda:")
costos_promedio = []
for i in range(len(tiendas)):
    # Sumar todos los costos de envío
    suma = 0
    conteo = 0
    for costo in tiendas[i]['Costo de envío']:
        suma = suma + costo
        conteo = conteo + 1
    # Calcular promedio
    promedio = suma / conteo
    costos_promedio.append(promedio)
    print(f"{nombres_tiendas[i]}: ${promedio:.2f}")
```
